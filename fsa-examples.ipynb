```python
import pandas as pd

# Get today's date
today = pd.Timestamp.today()

yesterday = pd.Timestamp.today() - pd.Timedelta(days=1)

yesterday = pd.Timestamp.today() - pd.Timedelta(days=1)
print(yesterday.strftime("%Y-%m-%d"))

# Find the previous Sunday's date
last_sunday = today - pd.Timedelta(days=today.weekday() + 1)

# Get the full week (Sunday to Saturday)
week_range = pd.date_range(start=last_sunday, periods=7, freq="D")

# Format the dates as YYYY-MM-DD
week_dates = week_range.strftime("%Y-%m-%d").tolist()

print(week_dates)

# Maybe run a job at 6am each day to add the previous day's items?
```
